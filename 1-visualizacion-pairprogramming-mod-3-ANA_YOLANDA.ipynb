```python
# 4. ¿Cuál es la mediana de la esperanza de vida en Europa? 
# Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.
```

```python
europa_df = df[df['continent'] == 'Europe']
```

```python
# Calcular la mediana de la esperanza de vida en Europa
mediana_esperanza = europa_df['life_expectancy'].median()
print(f"La mediana de la esperanza de vida en Europa es de {mediana_esperanza:.2f} años.")
```

```python
# Calcular el rango intercuartil (IQR)
q1 = europa_df['life_expectancy'].quantile(0.25)
q3 = europa_df['life_expectancy'].quantile(0.75)
iqr = q3 - q1
print(f"El rango intercuartil (IQR) es de {iqr:.2f} años.")
```

```python
# Determinar los límites para detectar valores atípicos (outliers)
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr
```

```python
# Identificar valores atípicos
valores_atipicos = europa_df[(europa_df['life_expectancy'] < limite_inferior) | 
                             (europa_df['life_expectancy'] > limite_superior)]

print(f"Número de valores atípicos: {len(valores_atipicos)}")
```

```python
sns.boxplot(x = europa_df['life_expectancy'], color='skyblue')
plt.title('Boxplot de la Esperanza de Vida en Europa')
plt.xlabel('Esperanza de Vida (años)')
```

```python
# 5. ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: 
# tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. 
# Describe cualquier patrón que encuentres.
fig, axes = plt.subplots(2, 2, figsize=(24, 20))
# Definir los pares de variables a comparar
variables = [
    ('birth_rate', 'infant_mortality'),
    ('birth_rate', 'life_expectancy'),
    ('birth_rate', 'density'),
    ('birth_rate', 'urban_population')
]


for ax, (x_var, y_var) in zip(axes.flatten(), variables):
    sns.scatterplot(data=df, x=x_var, y=y_var, ax=ax, color='blue')
    ax.set_title(f'{x_var} vs {y_var}', fontsize=14)
    ax.set_xlabel(x_var, fontsize=8)
    ax.set_ylabel(y_var, fontsize=8)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (40, 20))
sns.scatterplot(x = "birth_rate", y = 'infant_mortality', data = df, ax = axes[0][0])
sns.scatterplot(x = "birth_rate", y = 'life_expectancy', data = df, ax = axes[0][1])
sns.scatterplot(x = "birth_rate", y = 'density', data = df, ax = axes[1][0])
sns.scatterplot(x = "birth_rate", y = 'urban_population', data = df, ax = axes[1][1])
# ponemos titulos a las gráficas
axes[0][0].set_title("tasa de natalidad vs mortalidad infantil")
axes[0][1].set_title("tasa de natalidad vs esperanza de vida")
axes[1][0].set_title("tasa de natalidad vs densidad")
axes[1][1].set_title("tasa de natalidad vs población urbana")
# cambiamos nombres de los ejes
axes[0][0].set_xlabel("Tasa de natalidad")
axes[0][1].set_xlabel("Tasa de natalidad")
axes[1][0].set_xlabel("Tasa de natalidad")
axes[1][1].set_xlabel("Tasa de natalidad")
axes[0][0].set_ylabel("Mortalidad infantil")
axes[0][1].set_ylabel("Esperanza de vida")
axes[1][0].set_ylabel("Densidad")
axes[1][1].set_ylabel("Poblacion urbana");
```

```python
# 6. ¿Cuáles son los 10 países con la mayor área forestal?

# Asegúrate de que 'forest_area' está en el DataFrame y es numérico
df['forested_area'] = pd.to_numeric(df['forested_area'], errors='coerce')
```

```python
# Ordenar el DataFrame por la columna 'forest_area' de mayor a menor
top_10_forests = df[['country', 'forested_area']].sort_values(by='forested_area', ascending=False).head(10)

# Mostrar los 10 países con la mayor área forestal
print(top_10_forests)
```

```python
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
sns.barplot(x = 'forested_area', y = 'country', data = top_10_forests, palette = 'Blues')

# Configurar el título y etiquetas
plt.title('Top 10 Países con la Mayor Área Forestal', fontsize=16)
plt.xlabel('Área Forestal (km²)', fontsize=12)
plt.ylabel('País', fontsize=12)
```
