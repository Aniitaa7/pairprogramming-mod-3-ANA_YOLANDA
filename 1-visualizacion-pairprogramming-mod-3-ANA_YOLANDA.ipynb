```python
# 4. ¿Cuál es la mediana de la esperanza de vida en Europa? 
# Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.
```

```python
europa_df = df[df['continent'] == 'Europe']
```

```python
# Calcular la mediana de la esperanza de vida en Europa
mediana_esperanza = europa_df['life_expectancy'].median()
print(f"La mediana de la esperanza de vida en Europa es de {mediana_esperanza:.2f} años.")
```

```python
# Calcular el rango intercuartil (IQR)
q1 = europa_df['life_expectancy'].quantile(0.25)
q3 = europa_df['life_expectancy'].quantile(0.75)
iqr = q3 - q1
print(f"El rango intercuartil (IQR) es de {iqr:.2f} años.")
```

```python
# Determinar los límites para detectar valores atípicos (outliers)
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr
```

```python
# Identificar valores atípicos
valores_atipicos = europa_df[(europa_df['life_expectancy'] < limite_inferior) | 
                             (europa_df['life_expectancy'] > limite_superior)]

print(f"Número de valores atípicos: {len(valores_atipicos)}")
```

```python
sns.boxplot(x = europa_df['life_expectancy'], color='skyblue')
plt.title('Boxplot de la Esperanza de Vida en Europa')
plt.xlabel('Esperanza de Vida (años)')
```

```python
# 5. ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: 
# tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. 
# Describe cualquier patrón que encuentres.
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize = (20, 5))
```

```python
# Crear una figura con varios subgráficos (2 filas, 2 columnas)
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Gráfico 1: Relación entre la tasa de natalidad y la tasa de mortalidad infantil
sns.scatterplot(x='infant_mortality', y='rate_birth', data=df, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Tasa de Natalidad vs. Tasa de Mortalidad Infantil')
axes[0, 0].set_xlabel('Tasa de Mortalidad Infantil (por 1000 nacidos vivos)')
axes[0, 0].set_ylabel('Tasa de Natalidad (por 1000 habitantes)')

# Gráfico 2: Relación entre la tasa de natalidad y la esperanza de vida
sns.scatterplot(x='life_expectancy', y='rate_birth', data=df, ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Tasa de Natalidad vs. Esperanza de Vida')
axes[0, 1].set_xlabel('Esperanza de Vida (años)')
axes[0, 1].set_ylabel('Tasa de Natalidad (por 1000 habitantes)')

# Gráfico 3: Relación entre la tasa de natalidad y la densidad de población
sns.scatterplot(x='density', y='rate_birth', data=df, ax=axes[1, 0], color='salmon')
axes[1, 0].set_title('Tasa de Natalidad vs. Densidad de Población')
axes[1, 0].set_xlabel('Densidad de Población (hab/km²)')
axes[1, 0].set_ylabel('Tasa de Natalidad (por 1000 habitantes)')

# Gráfico 4: Relación entre la tasa de natalidad y la población urbana
sns.scatterplot(x='urban_population', y='rate_birth', data=df, ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Tasa de Natalidad vs. Población Urbana')
axes[1, 1].set_xlabel('Población Urbana (%)')
axes[1, 1].set_ylabel('Tasa de Natalidad (por 1000 habitantes)')
```
